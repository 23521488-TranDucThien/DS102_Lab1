```python
# Hàm chuyển tháng sang số
def convert_month(month: str) -> int:
    month_dict = {
        'jan': 1,
        'feb': 2,
        'mar': 3,
        'apr': 4,
        'may': 5,
        'jun': 6,
        'jul': 7,
        'aug': 8,
        'sep': 9,
        'oct': 10,
        'nov': 11,
        'dec': 12
    }
    return month_dict[month]
```

```python
def convert_day(day: str) -> int:
    day_dict = {
        'sun': 0,
        'mon': 1, 
        'tue': 2,
        'wed': 3,
        'thu': 4,
        'fri': 5,
        'sat': 6,
    }
    return day_dict[day]
```

```python
# Chuyển chữ sang số
df['day'] = df['day'].apply(convert_day)
```

```python
# Chuyển chữ sang số
df['month'] = df['month'].apply(convert_month)
```

```python
# Phân Tích Phân Phối Dữ Liệu
plt.figure(figsize=(8, 6))
sns.histplot(df['area'], bins=50, kde=True)
plt.title('Phân phối diện tích cháy rừng')
plt.show()
```

```python
# Phát Hiện Giá Trị Ngoại Lai
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain']])
```

```python
# Phân tích tương quan giữa các biến
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

```python
# Danh sách các cột cần phân tích
features = ['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain']

# Vẽ scatter plot cho từng biến với 'area'
plt.figure(figsize=(10, 15))
for i, feature in enumerate(features, 1):
    plt.subplot(4, 2, i)
    sns.scatterplot(data=df, x=feature, y='area', alpha=0.6)
    plt.title(f'{feature} vs Area')

plt.tight_layout()
plt.show()
```

```python
X = df.iloc[:, :-1].values 
y = df.iloc[:, -1].values.reshape(-1, 1)

np.random.seed(42)
indices = np.random.permutation(len(X)) 

train_size = int(0.8 * len(X))  

X_train, X_test = X[indices[:train_size]], X[indices[train_size:]]
y_train, y_test = y[indices[:train_size]], y[indices[train_size:]]

print(f"Train size: {X_train.shape}, Test size: {X_test.shape}")
```

```python
from tqdm import tqdm
```

```python
class SoftmaxRegression:
    def __init__(self, epochs: int, lr: float) -> None:
        self.epochs = epochs
        self.lr = lr
        self.losses = []
        self.accuracies = []
        self.theta = None

    @staticmethod
    def softmax(z: np.ndarray) -> np.ndarray:
        exp_z = np.exp(z - np.max(z, axis=1, keepdims=True)) 
        return exp_z / np.sum(exp_z, axis=1, keepdims=True)

    @staticmethod
    def loss_fn(y: np.ndarray, y_hat: np.ndarray) -> float:
        return -np.mean(np.sum(y * np.log(y_hat + 1e-6), axis=1))

    @staticmethod
    def accuracy(y: np.ndarray, y_hat: np.ndarray) -> float:
        return np.mean(np.argmax(y, axis=1) == np.argmax(y_hat, axis=1))

    def fit(self, X: np.ndarray, y: np.ndarray) -> None:
        n_samples, n_features = X.shape
        n_classes = y.shape[1]
        self.theta = np.zeros((n_features, n_classes))

        for epoch in tqdm(range(self.epochs), desc="Training Progress", leave=True):
            logits = X @ self.theta
            y_hat = self.softmax(logits)

            loss = self.loss_fn(y, y_hat)
            acc = self.accuracy(y, y_hat)

            self.losses.append(loss)
            self.accuracies.append(acc)

            grad = (1 / n_samples) * X.T @ (y_hat - y)
            self.theta -= self.lr * grad
    
    def predict(self, X: np.ndarray) -> np.ndarray:
        logits = X @ self.theta
        return self.softmax(logits)

    def predict_classes(self, X: np.ndarray) -> np.ndarray:
        return np.argmax(self.predict(X), axis=1)
```

```python
model = SoftmaxRegression(epochs=1000, lr=0.1)

model.fit(X_train, y_train)

y_pred = model.predict_classes(X_test)

test_accuracy = model.accuracy(y_test, np.eye(y_test.shape[1])[y_pred])
print(f"Test Accuracy: {test_accuracy:.4f}")
```
